```python
import numpy as np
import matplotlib.pyplot as plt

# 두 직선의 계수를 정의합니다
# 첫 번째 직선: 2x + 3y = 5
# 두 번째 직선: -x + y = 5
A = np.array([[2, 3], [-1, 1]])
B = np.array([5, 5])

# 교점 계산
intersection = np.linalg.solve(A, B) #선형대수 (linear algebra)의 해를 구한다(solve)
x_intersection, y_intersection = intersection

# 직선을 그리기 위해 x 범위 설정
x_values = np.linspace(-10, 10, 100)

# 두 직선의 y 값 계산
y1_values = (5 - 2 * x_values) / 3
y2_values = 5 + x_values

# 시각화
plt.figure(figsize=(8, 6))
plt.plot(x_values, y1_values, label="2x + 3y = 5")
plt.plot(x_values, y2_values, label="-x + y = 5")
plt.scatter(x_intersection, y_intersection, color='red', zorder=5, label=f'Intersection ({x_intersection:.2f}, {y_intersection:.2f})')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Intersection of Two Lines")
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 계수 행렬 C와 상수 벡터 D 정의
C = np.array([[3, 2, 6], [-2, 4, 2], [6, 7, 3]])
D = np.array([9, 11, 2])

# 교점 계산
intersection = np.linalg.solve(C, D)

# x, y, z 값으로 unpack
x_intersection, y_intersection, z_intersection = intersection

# 결과 출력
print(f"Intersection: x = {x_intersection}, y = {y_intersection}, z = {z_intersection}")

# 시각화용 x, y 값 계산
# 직선을 그리려면 2차원으로 줄여야 하기 때문에, x, y 평면을 그리겠습니다.
# (예시로 특정 범위에서 x 값에 대해 y를 계산하는 방식)

# 시각화용 x 범위 설정
x_values = np.linspace(-10, 10, 100)

# 두 직선의 방정식을 y 값으로 변환하여 그리기
y1_values = (9 - 3 * x_values) / 2
y2_values = (11 - 4 * x_values) / 2

# 시각화
plt.figure(figsize=(8, 6))

# 두 직선 그리기
plt.plot(x_values, y1_values, label="3x + 2y + 6z = 9")
plt.plot(x_values, y2_values, label="-2x + 4y + 2z = 11")

# 교차점 시각화
plt.scatter(x_intersection, y_intersection, color='red', zorder=5, label=f'Intersection ({x_intersection:.2f}, {y_intersection:.2f}, {z_intersection:.2f})')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Intersection of Two Lines (2D Projection)")
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# 세 개의 3차원 벡터 정의
A_three_dim = np.array([1, 0, 2])
B_three_dim = np.array([0, 1, 3])
C_three_dim = np.array([1, 1, 1])

# 목표 벡터
target_vector = np.array([9, 11, 2])

# 선형 결합 계수 계산
coeff_matrix = np.column_stack([A_three_dim, B_three_dim, C_three_dim])
intersection_b = np.linalg.solve(coeff_matrix, target_vector)

# zip을 이용하여 선형 결합 계산
linear_combination = sum(i * j for i, j in zip([A_three_dim, B_three_dim, C_three_dim], intersection_b))

# 출력
print("계산된 선형 결합 계수 (intersection_b):", intersection_b)
print("선형 결합 결과:", linear_combination)
print("원래 목표 벡터:", target_vector)
print("일치 여부:", np.allclose(linear_combination, target_vector))
```
